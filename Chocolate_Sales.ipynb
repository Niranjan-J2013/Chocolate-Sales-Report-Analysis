```python
import pandas as pd
df=pd.read_csv('Chocolate Sales-2.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe(include='all')
```

```python
# checking null values
df.isnull().sum()
```

```python
df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace(' ','_')
```

```python
df.columns
```

```python
# Convert Amount column to numeric
df["amount"] = df["amount"].replace('[\$,]', '', regex=True).astype(float)
```

```python
# Convert Date column
df["date"] = pd.to_datetime(df["date"], dayfirst=True)
```

```python
df.info()
```

```python
# creating month column for dashborad
df["month"] = df["date"].dt.to_period("M")
```

```python
df.head()
```

```python
#performing sales anaysis
#1. Total sales:
df["amount"].sum()
```

```python
#2. sales by country:
df.groupby("country")["amount"].sum().sort_values(ascending=False)
```

```python
#3. Top products:
df.groupby("product")["amount"].sum().sort_values(ascending=False)
```

```python
#4. Top sales person:
df.groupby("sales_person")["amount"].sum().sort_values(ascending=False)
```

```python
#4.1. Top 10 sales person:
df.groupby("sales_person")["amount"].sum().sort_values(ascending=False).head(10)
```

```python
#5. Boxes shipped by country:
df.groupby("country")["boxes_shipped"].sum().sort_values(ascending=False)
```

```python
#5.1. Top 3 Boxes shipped by country:
df.groupby("country")["boxes_shipped"].sum().sort_values(ascending=False).head(3)
```

```python
#Export data for powerbi
df.to_csv("Chocolate_sales_dashboard_dataset.csv", index=False)
```
